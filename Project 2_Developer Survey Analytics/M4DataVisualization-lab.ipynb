## Demo: How to work with database

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT Age,WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35 ;

"""
df_stack =pd.read_sql_query(QUERY,conn)
df_stack

```

```python
import matplotlib.pyplot as plt
import numpy as np

med_WorkWeekHrs = df_stack['WorkWeekHrs'].median()
med_CodeRevHrs = df_stack['CodeRevHrs'].median()

max_WorkWeekHrs = df_stack['WorkWeekHrs'].max()
max_CodeRevHrs = df_stack['CodeRevHrs'].max()

#y_below_WorkWeekhrs = df[(df['WorkWeekHrs'] <= med_WorkWeekHrs)].count()
#y_below_CodeRevHrs = df[(df['CodeRevHrs'] <= med_CodeRevHrs)].count()

print(med_CodeRevHrs)


# create data
x = ['WorkWeekHrs', 'odeRevHrs']
y_below = np.array([med_WorkWeekHrs,med_CodeRevHrs])
y_above = np.array([max_WorkWeekHrs,max_CodeRevHrs])

# plot bars in stack manner
plt.bar(x, y_below, color='r')
plt.bar(x, y_above, bottom=y_below, color='b')

plt.xlabel("Column")
plt.ylabel("Value")
plt.legend(["Below med", "Above med"])
plt.title("Stacked chart")
plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python

QUERY = """
SELECT Age,ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60 
ORDER BY Age;

"""
df_line =pd.read_sql_query(QUERY,conn)
df_line
```

```python
med_list =[]
for i in range (45,61) : 
    df_line_age= df_line[(df_line['Age'] == i)]
    med_ConvertedComp_age = df_line_age['ConvertedComp'].median()
    med_list.append(med_ConvertedComp_age)
print(med_list)
```

```python

# importing the required libraries 
import matplotlib.pyplot as plt 
import numpy as np 
  
# define data values 
x = np.array(list(range(45, 61)))  # X-axis points 
y = np.array(med_list)  # Y-axis points 
  
plt.plot(x, y)  # Plot the chart 
plt.show()  # display
```

```python
df_x['Age'] = pd.DataFrame(x)
df_y['Med'] = pd.DataFrame(y)

corr = df_x['Age'].corr(df_y['Med'])
print(corr)
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python

QUERY = """
SELECT MainBranch
FROM master

"""
df_bar =pd.read_sql_query(QUERY,conn)
df_bar.value_counts()
```

```python

from matplotlib import pyplot as plt
 
 
# Figure Size
barWidth = 0.25
#fig = plt.figure(figsize =(10, 7))

# set height of bar 
bar_height = [df_bar.value_counts()[0], df_bar.value_counts()[1]]
print(bar_height)
# Set position of bar on X axis 
bar_width_1 = np.arange(len(bar_height)) # [0,1,2]
bar_width_2 = [x + barWidth for x in bar_width_1] 
 
# Make the plot
plt.bar(bar_width_1, bar_height[0], color ='r', width = barWidth, 
        edgecolor ='grey', label ='I am a developer by profession  ') 
plt.bar(bar_width_2, bar_height[1], color ='g', width = barWidth, 
        edgecolor ='grey', label ='I am not primarily a developer, but I write code sometimes as part of my work')
 
# Show Plot
plt.show()
```
